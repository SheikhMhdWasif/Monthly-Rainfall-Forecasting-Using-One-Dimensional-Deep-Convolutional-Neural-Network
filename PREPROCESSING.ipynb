```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Source_Data_ .csv")
```

```python
data.tail(32)
```

```python
data.drop(['Product code','Bureau of Meteorology station number','Day','Period over which rainfall was measured (days)','Quality'],axis = 1,inplace = True)
```

```python
condition = (data['Year'] >= 1957)
data = data[condition]
data.reset_index(drop=True, inplace=True)
```

```python
#Checking for missing values or NAN Data.
missing_data = data.isnull()
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")
```

```python
avg = data["Rainfall amount (millimetres)"].astype('float').mean(axis=0)
print("X:", avg)
```

```python
data["Rainfall amount (millimetres)"].replace(np.nan, avg,inplace = True)
```

```python
#Checking for missing values or NAN Data.
missing_data = data.isnull()
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")
```

```python
data.to_csv("Rainfall_Data.csv", index=False)
```
